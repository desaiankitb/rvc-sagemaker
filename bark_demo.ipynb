## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt = """
     Hey team, it's Josh. [clears throat] I just spent 30 minutes recording
     my voice samples for Genuin's AI project, and the process was straightforward.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

## Advanced examples

```python
text_prompt = """
    I believe in the potential of this technology and how it can revolutionize our work.
    If we all contribute our voice samples, it ensures our AI system will be more diversified,
    versatile, and ultimately, effective for our projects.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    If you haven't done so yet, please take out a small chunk of your day, just half an hour,
    to record your voice. Let's all be a part of this innovative step forward for Genuin.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```
